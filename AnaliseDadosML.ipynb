Este script Python extrai e analisa dados da API pública do Mercado Livre para a busca "cofre". As visualizações geradas se adaptam automaticamente a outros termos de busca.

**I. Extração dos dados**

```python
import requests
import pandas as pd

# Parâmetros da busca
termo = "cofre"
limite = 50

# Requisição à API
url = f"https://api.mercadolibre.com/sites/MLB/search?q={termo}&limit={limite}"
resposta = requests.get(url)

# Extração dos dados
dados = resposta.json()["results"]

# Conversão para DataFrame
df = pd.DataFrame(dados)

# Salvando o DataFrame
df.to_csv("dados_cofre.csv", index=False)
```

**II. Análise exploratória dos dados**

```python
# Importando bibliotecas
import matplotlib.pyplot as plt
import seaborn as sns

# Visualização da distribuição de preços
plt.figure(figsize=(10,6))
sns.displot(df.price, kind="kde")
plt.title("Distribuição de Preços")
plt.show()

# Visualização da quantidade de cofres por tipo de envio
plt.figure(figsize=(10,6))
sns.countplot(x=df.shipping.free_shipping, data=df)
plt.title("Quantidade de Cofres por Tipo de Envio")
plt.show()

# Visualização da correlação entre preço e quantidade vendida
plt.figure(figsize=(10,6))
sns.scatterplot(x=df.price, y=df.sold_quantity)
plt.title("Correlação entre Preço e Quantidade Vendida")
plt.show()
```

**III. Considerações sobre os dados**

A maioria dos cofres tem preço entre 100 e 500 R$.
Há uma quantidade maior de cofres com frete grátis.

Existe uma correlação positiva entre preço e quantidade vendida, ou seja, cofres mais caros tendem a vender mais.
